MEDICAL INSURANCE COST PREDICION

About the Dataset-

- age: age of primary beneficiary

- sex: insurance contractor gender, female, male

- bmi: Body mass index, providing an understanding of body, weights that are relatively high or low relative to height,
  objective index of body weight (kg / m ^ 2) using the ratio of height to weight, ideally 18.5 to 24.9

- children: Number of children covered by health insurance / Number of dependents

- smoker: Smoking

- region: the beneficiary's residential area in the US, northeast, southeast, southwest, northwest.

- charges: Individual medical costs billed by health insurance

```python
#importing the dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.preprocessing import LabelEncoder
```

```python
#importing the dataset into a dataframe
data = pd.read_csv('insurance.csv')
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
#checking for null values
data.isnull().sum()
```

```python
data.sex.value_counts()
```

```python
data.region.value_counts()
```

```python
encoder = LabelEncoder()
```

```python
labels = encoder.fit_transform(data.sex)
```

```python
data['sex'] = labels
```

```python
data.head()
```

SEX

- MALE -> 1
- FEMALE -> 0

```python
labels = encoder.fit_transform(data.region)
```

```python
data['region'] = labels
```

```python
data.head()
```

REGION

- Northeast -> 0
- Northwest -> 1
- Southeast -> 2
- Southwest -> 3

```python
labels = encoder.fit_transform(data.smoker)
```

```python
data['smoker'] = labels
```

```python
data.head()
```

SMOKER

- Yes -> 1
- No -> 0

```python
#splitting the parameters and target variable
X = data.drop(columns='charges',axis=1)
Y = data['charges']
```

```python
#train test split the dataset
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state=101)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
model = RandomForestRegressor()
```

```python
model.fit(X_train,Y_train)
```

```python
testing_data_prediction = model.predict(X_test)
```

```python
score = metrics.r2_score(Y_test,testing_data_prediction)
```

```python
score
```

```python
#BULDING THE PREDICTION SYSTEM
input_data = (19,0,27.9,0,1,3)
input_data_array = np.asarray(input_data)

#reshaping the data so that it works for only one instance at a time
input_data_reshaped = input_data_array.reshape(1,-1)

prediction = model.predict(input_data_reshaped)
print('Predicted Medical Insurance Cost : ',str(prediction))

```
